```python
# basado en el LIBRO de estadistica para administracion pagina 12
#creamos una lista vacia que representa la edad de 100 personas
import random
edades = []
# hacemos un ciclo para generar 100 numeros aleatorios que representen la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  # num_aleatorio
  numero_aleatorio = random.randint(18,70)
  # agregamos el numero aleatorio a la lista edades
  edades.append(numero_aleatorio)
  print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los numeros aleatorios
df= pd.DataFrame(edades, columns=['Edades'])
df.head()
```

```python
# Calcular la tabla de distribuci√≥n de frecuencias
tabla_frecuencias = df['Edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de distribuci√≥n de frecuencias
print("Tabla de distribuci√≥n de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
#importamos la libreria de funciones aritmeticas
import math
```

```python
#volvemos a usar la lista de edades
print(edades)
```

```python
# crear un DataFrame con la edades
df = pd.DataFrame({'edad': edades})
df
```

```python
# calcular el numero de intervalos utilizando la regla de Sturges
n = len(edades)
k = int( 1 + math.log2(n))
print("valor de n =", n, "nNumero de intervalos =", k)
```

```python
#crear la tabla de frecuencia
tabla_frecuencias = pd.cut(df['edad'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['clase', 'frecuencia']
tabla_frecuencias
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='clase')
display(tabla_frecuencias)
```

```python
# Genere\ar columnas siguientes
tabla_frecuencias ['Frec Rel'] = tabla_frecuencias ['frecuencia'] /n
tabla_frecuencias ['Frec Porc'] = tabla_frecuencias ['frecuencia'] / n * 100
tabla_frecuencias ['Frec Ac'] = tabla_frecuencias['frecuencia'].cumsum()
tabla_frecuencias ['Frec Rel Ac'] = tabla_frecuencias ['Frec Rel'].cumsum()
tabla_frecuencias [ 'Frec Porc Ac'] = tabla_frecuencias ['Frec Porc'].cumsum()
# Mostrar la tabla de frecuencia
print(tabla_frecuencias)
```
