```python
import math
from math import comb

n = 1000
p = 0.001

# a) P(X = 0)
p0 = (1-p)**n
p0
```

```python
# b) P(X = 3)
p3 = comb(n,3) * p**3 * (1-p)**(n-3)
p3
```

```python
# c) P(X = 4)
p4 = comb(n,4) * p**4 * (1-p)**(n-4)
p4
```
