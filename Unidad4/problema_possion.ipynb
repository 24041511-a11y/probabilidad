```python
import math
from math import exp, factorial

# Parámetro lambda
lambda_poisson = 5

# a) P(X = 2)
x = 2
p_x_2 = exp(-lambda_poisson) * lambda_poisson**x / factorial(x)
p_x_2
```

```python
# b) P(X > 10)
import math

prob = 1 - sum(math.exp(-lambda_poisson) * lambda_poisson**k / math.factorial(k) for k in range(0, 11))
prob
```
