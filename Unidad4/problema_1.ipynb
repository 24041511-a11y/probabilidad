```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

media = 70
desv_est = 4
n = 500
x1 = 68
x2 = 72

p1 = norm.cdf(x1, media, desv_est)
p2 = norm.cdf(x2, media, desv_est)
prob_rango = p2 - p1
estudiantes_en_rango = prob_rango * n

print("============================================")
print("      Resultados del análisis de pesos       ")
print("============================================")
print(f"Media (μ): {media} kg")
print(f"Desviación estándar (σ): {desv_est} kg")
print(f"Rango considerado: {x1} kg a {x2} kg\n")
print(f"Probabilidad (68 < X < 72): {prob_rango:.4f}")
print(f"Número estimado de estudiantes: {estudiantes_en_rango:.2f} de {n}")
print("============================================")

x = np.linspace(media - 4*desv_est, media + 4*desv_est, 1000)
y = norm.pdf(x, media, desv_est)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Distribución Normal", color="blue")
plt.fill_between(x, y, where=(x >= x1) & (x <= x2), color="lightgreen", alpha=0.5, label=f"68 < X < 72")
plt.title("Distribución normal de los pesos (ITD)")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```
